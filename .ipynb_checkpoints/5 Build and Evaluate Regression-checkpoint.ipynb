# Define Task
"Make predictions about a country's life expectancy in years from a set of metrics for the country"

# Acquire Data

```python
dataset = pd.read_csv("./OC_csv/world_data.csv")

dataset = dataset.drop([ 'murder','urbanpopulation','unemployment'], axis=1)

print(dataset.dtypes)
print("\n\n : ",dataset.isnull().sum())
```

```python
numeric_means = dataset.select_dtypes(exclude=["object"]).mean().to_dict()

for m in numeric_means:
    dataset[m] = dataset[m].fillna(value= numeric_means[m])
dataset.isnull().sum()
```

# Prepare Data

```python
dataset.columns
```

```python
y = dataset["lifeexp"]
X = dataset[['happiness', 'income', 'sanitation', 'water',
       'literacy', 'inequality', 'energy', 'childmortality', 'fertility',
       'hiv', 'foodsupply', 'population']]
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
rescaled_X = scaler.fit_transform(X)
X = pd.DataFrame(rescaled_X, index=X.index, columns=X.columns)
```

```python
seed = 1
test_size = 0.33
X_train, X_test,Y_train, Y_test = train_test_split(X, y, random_state=seed, test_size= test_size)
```

```python
models = [LinearRegression(), KNeighborsRegressor(),SVR()]

for model in models:
    model.fit(X_train, Y_train)
    prediction = model.predict(X_train)
    print(type(model).__name__, mean_absolute_error(Y_train, prediction))
```

# Evaluate Model

```python
for model in models:
    model.fit(X_test,Y_test)
    prediction = model.predict(X_test)
    print(f"{type(model).__name__} : {mean_absolute_error(Y_test,prediction)}")
```
